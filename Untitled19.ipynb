```python
import random

def read_student_ids(file_name):
    try:
        with open(file_name, 'r') as file:
            student_ids = file.read().splitlines()
        return student_ids
    except FileNotFoundError:
        print(f"Error: File '{file_name}' not found.")
        return []

def initialize_students(file_name):
    student_ids = read_student_ids(file_name)
    return student_ids

def select_student(students):
    selected_student = random.choice(students)
    students.remove(selected_student)
    return selected_student

def conduct_viva(file_name):
    while True:
        students = initialize_students(file_name)
        if not students:
            return

        viva_count = 1
        while students:
            selected_student = select_student(students)
            print(f"Viva #{viva_count}: Student selected - {selected_student}")
            viva_count += 1

        print("All students have been selected. Resetting list.\n")

# Example usage
conduct_viva('ID.txt')
```
